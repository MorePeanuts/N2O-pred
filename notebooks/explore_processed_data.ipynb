```python
import pickle
import random
import numpy as np
```

```python
encoders_path = '../datasets/encoders.pkl'
with open(encoders_path, 'rb') as f:
    encoders = pickle.load(f)

data_path = '../datasets/data_EUR_processed.pkl'
with open(data_path, 'rb') as f:
    data = pickle.load(f)
```

```python
seq_data = random.choice(data)
seq_data
```

```python
CATEGORICAL_STATIC_FEATURES = ['crop_class']
CATEGORICAL_DYNAMIC_FEATURES = ['fertilization_class', 'appl_class']

categorical_static_feature = np.array(seq_data['categorical_static'])
encoders['crop_class'].inverse_transform(categorical_static_feature)
```

```python
categorical_dynamic_feature = np.array(seq_data['categorical_dynamic'])
encoders['fertilization_class'].inverse_transform(categorical_dynamic_feature[:, 0])
```

```python
encoders['appl_class'].inverse_transform(categorical_dynamic_feature[:, 1])
```
