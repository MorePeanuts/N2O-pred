```python
import pickle
import random
import numpy as np
from n2o_pred.data import (
    CATEGORICAL_DYNAMIC_FEATURES,
    CATEGORICAL_STATIC_FEATURES,
)
```

```python
encoders_path = '../datasets/encoders.pkl'
with open(encoders_path, 'rb') as f:
    encoders = pickle.load(f)

data_path = '../datasets/data_EUR_processed.pkl'
with open(data_path, 'rb') as f:
    data = pickle.load(f)
```

```python
seq_data = random.choice(data)
seq_data
```

```python
categorical_static_feature = np.array(seq_data['categorical_static'])
encoders[CATEGORICAL_STATIC_FEATURES[0]].inverse_transform(categorical_static_feature)
```

```python
categorical_dynamic_feature = np.array(seq_data['categorical_dynamic'])
encoders[CATEGORICAL_DYNAMIC_FEATURES[0]].inverse_transform(categorical_dynamic_feature[:, 0])
```
