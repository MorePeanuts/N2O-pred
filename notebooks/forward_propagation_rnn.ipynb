```python
import random
import torch
from pprint import pprint
from pathlib import Path
from n2o_pred.models import RNNModel
from n2o_pred.data import SequentialN2ODataset
```

```python
dataset = SequentialN2ODataset(Path('../datasets/data_EUR_processed.pkl'))
seq_data = random.choice(dataset)
pprint(seq_data.to_dict())
```

```python
model = RNNModel(
    num_numeric_static=dataset.get_num_numeric_static(),
    num_numeric_dynamic=dataset.get_num_numeric_dynamic(),
    categorical_static_cardinalities=dataset.get_categorical_static_cardinalities(),
    categorical_dynamic_cardinalities=dataset.get_categorical_dynamic_cardinalities(),
)

output = model.forward(
    numeric_static_features=torch.tensor(seq_data.numeric_static).unsqueeze(0),
    numeric_dynamic_features=torch.tensor(seq_data.numeric_dynamic)[:, 0:6].unsqueeze(0),
    categorical_static_features=torch.tensor(seq_data.categorical_static).unsqueeze(0),
    categorical_dynamic_features=torch.tensor(seq_data.categorical_dynamic).unsqueeze(0),
    seq_lengths=torch.tensor([seq_data.seq_length]),
)
```
